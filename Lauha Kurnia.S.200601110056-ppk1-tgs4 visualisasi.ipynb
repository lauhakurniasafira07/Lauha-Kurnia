#### 2. Cari bagaimana cara memberikan label pada sumbu x dan sumbu y

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[44,46,47,45,47,49,48]
plt.plot(x,y)

plt.title('Program Menggemukan Badan')
plt.xlabel('Tanggal')
plt.ylabel('Berat Badan')

plt.grid(True)
plt.show()
```

#### 3. Cari bagaimana cara menambahkan teks dalam grafik

```python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[44,46,47,45,47,49,48]
plt.plot(x,y)

plt.title('Program Menggemukan Badan')
plt.xlabel('Tanggal')
plt.ylabel('Berat Badan')
plt.text(3,47,"Berat Badan Mengalami Penurunan pada Tanggal 4 dan 7")


plt.grid(True)
plt.show()
```
